```python
product_name=[]
prices=[]
description=[]
reviews=[]


for i in range(1,12):
    url="https://www.flipkart.com/search?q=phones+under+10000&sid=tyy%2C4io&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_12_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_12_na_na_na&as-pos=1&as-type=RECENT&suggestionId=phones+under+10000%7CMobiles&requestId=f7369335-5cf3-46e3-896e-549203c72a63&as-backfill=on&page="+str(i)
    r=requests.get(url)
    soup=bs(r.text,"lxml")

    names=soup.find_all("div",class_="RG5Slk")
    print(names)


    for i in names:
        name=i.text
        product_name.append(name)
    print(product_name)


    pp=soup.find_all("div",class_="hZ3P6w DeU9vF")

    for i in pp:
        name=i.text
        prices.append(name)

    print(prices)

    des=soup.find_all("ul",class_="HwRTzP")

    for i in des:
        name=i.text
        description.append(name)
        
    print(description)

    rev=soup.find_all("div",class_="MKiFS6")
    for i in rev:
        name=i.text
        reviews.append(name)
        
        
    print(reviews)
    print(len(reviews))

min_len = min(len(product_name), len(prices), len(description), len(reviews))
product_name = product_name[:min_len]
prices = prices[:min_len]
description = description[:min_len]
reviews = reviews[:min_len]

df = pd.DataFrame({
    "Product Name": product_name,
    "Price": prices,
    "Description": description,
    "Rating": reviews
})



print(df.head())

df.to_csv("flipkart_phones_under_10000.csv", index=False)


# np=soup.find("a",class_="i2eZXn").get("href")
# cmp="https://www.flipkart.com"+np
# print(cmp)
```
